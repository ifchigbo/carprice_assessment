```python
import tui
from getDatPDDF import getCarasPDDF
import _1_Car_Record_byID
import _2_Car_Record_Cylinder_number
import _3_Car_Spec_Body_type
import _4_Column_Data_Per_Car
import _5_Car_Name_Alfa
import _6_Summary_Sales
import _7_Top5_Sales_Price
import _8_Car_DetailsBySelection
import _9_Fuel_Consumption_Horse_power
import _10_Buying_Behavior
import loaddata



def run():
    
    try:

        tui.startModule("Started Loading CSV file as Pandas Data Frame ......................")
        print("\n")
        df = getCarasPDDF()
        print("\n")
        tui.completed("Completed Loading CSV file as Pandas Data Frame.")

        while True:
            myselection = tui.programmenu()
            if myselection == 0:
                break
            elif myselection == 1:
                tui.startModule('Listing Car By ID Module Started')
                print("\n")
                _1_Car_Record_byID.displayCarDataQuerybyID()
                print("\n")
                tui.completed("Listing Car by ID Module Completed")
            elif myselection == 2: 
                tui.startModule("Listing Cars by Cylinder Number")
                print("\n")
                _2_Car_Record_Cylinder_number.getCarbyCylinderNumber(df)
                print("\n")
                tui.completed("Listing Cars by Cylinder number completed")
            elif myselection ==3:
                tui.startModule("Listing Cars by Specific Body Type")
                print("\n")
                _3_Car_Spec_Body_type.retrieveCarsBody(df)
                print("\n")
                tui.completed("Listing Cars by Specific Body Type Completed")
            elif myselection == 4:
                tui.startModule("Listing Columns of your specified Car by ID")
                print("\n")
                _4_Column_Data_Per_Car.getSpecColumnsbyCarID(df)
                print("\n")
                tui.completed("Listing Columns of your specified Car by ID Completed")
            elif myselection == 5:
                tui.startModule("Listing Car Names in Alphabetical Order")
                print("\n")
                _5_Car_Name_Alfa.sortCarbyAlph()
                print("\n")
                tui.startModule("Listing Car Names in Alphabetical Order Completed")
            elif myselection == 6:
                tui.startModule("Listing Summary of Sales for each Car Body")
                print("\n")
                _6_Summary_Sales.getSalesSummary(df)
                print("\n")
                tui.completed("Listing Summary of Sales for each Car Body Completed")
            elif myselection == 7:
                tui.startModule("Listing Top 5 Selling Cars by Price")
                print("\n")
                _7_Top5_Sales_Price.getMostSellingCars(df)
                print("\n")
                tui.completed("Module listing Completed")
            elif myselection == 8:
                tui.startModule("Listing Car Details by User Selection")
                print("\n")
                _8_Car_DetailsBySelection.getSpecColumnsbyCarID(df)
                print("\n")
                tui.completed("Listing Car Details by User Selection Completed")
            elif myselection == 9:
                tui.startModule("Getting Sub Plot for Car Fuel System and Horse Power of Top 5 Cars by Sales  ")
                print("\n")
                _9_Fuel_Consumption_Horse_power.getBarChart()
                print("\n")
                tui.completed("Completed Subplot Listing.")
            elif myselection == 10:
                tui.startModule("Getting Sub Plots for Decision Analysis ")
                print("\n")
                _10_Buying_Behavior.getAnalytics()
                print("\n")
                tui.completed("Completed Decsion Analysis.")
            else:
                tui.errormessage()
                
    except BaseException as error:
        print("The following error has occrued : {}".format(error))

if __name__ == '__main__':
    run()
    




```
